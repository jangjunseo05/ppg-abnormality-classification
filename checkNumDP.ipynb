```python
import numpy as np
import os
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# PPG ë°ì´í„° ë¡œë“œ
ppg_data = np.load(r"C:\Users\USER\Desktop\PPG\PPG_graph\npy_test\2\sample_2.npy")

# í”¼í¬ ì°¾ê¸°
peaks, _ = find_peaks(ppg_data, distance=116) 

# í”¼í¬ ê°œìˆ˜ë¥¼ 10~25ê°œ ë‹¨ìœ„ë¡œ ë³€ê²½í•˜ë©° ìµœëŒ€ ë°ì´í„° í¬ì¸íŠ¸ ê°œìˆ˜ ë¶„ì„
results = {}

for n in range(8, 26):  # ğŸ”¹ í”¼í¬ ê°œìˆ˜ë¥¼ 8~25ê°œê¹Œì§€ ë³€í™”
    segment_lengths = []

    for i in range(0, len(peaks) - (n - 1), n):  # nê°œ í”¼í¬ ë‹¨ìœ„ë¡œ ìŠ¬ë¼ì´ë”©
        start_idx = peaks[i]  # ì‹œì‘ì  (ì²« ë²ˆì§¸ í”¼í¬ ìœ„ì¹˜)
        end_idx = peaks[i + (n - 1)]  # në²ˆì§¸ í”¼í¬ ìœ„ì¹˜
        segment = ppg_data[start_idx:end_idx]  # í•´ë‹¹ êµ¬ê°„ ë°ì´í„° ì¶”ì¶œ
        
        segment_lengths.append(len(segment))  # í•´ë‹¹ ì„¸ê·¸ë¨¼íŠ¸ì˜ ë°ì´í„° í¬ì¸íŠ¸ ê°œìˆ˜ ì €ì¥

    # ìµœëŒ€ ê¸¸ì´ ì €ì¥
    if segment_lengths:
        max_length = np.max(segment_lengths)
    else:
        max_length = 0  # ë°ì´í„°ê°€ ì—†ìœ¼ë©´ 0

    results[n] = max_length  # í”¼í¬ ê°œìˆ˜(n)ë³„ ìµœëŒ€ ë°ì´í„° í¬ì¸íŠ¸ ê°œìˆ˜ ì €ì¥

# ê²°ê³¼ ì¶œë ¥
print("âœ… í”¼í¬ ê°œìˆ˜ë³„ ìµœëŒ€ ë°ì´í„° í¬ì¸íŠ¸ ê°œìˆ˜:")
for n, max_len in results.items():
    print(f"ğŸ”¹ í”¼í¬ {n}ê°œ ë‹¨ìœ„ â†’ ìµœëŒ€ ê¸¸ì´: {max_len}")

# ë°ì´í„° ê¸¸ì´ ë³€í™” ì‹œê°í™”
plt.figure(figsize=(10, 5))
plt.plot(list(results.keys()), list(results.values()), marker='o', linestyle='-', color='blue', alpha=0.8)
plt.xlabel("Number of Peaks per Segment")
plt.ylabel("Max Segment Length (Data Points)")
plt.title("Max PPG Segment Lengths for Different Peak Counts")
plt.grid(True)
plt.xticks(range(10, 26))  # xì¶• í”¼í¬ ê°œìˆ˜ ë²”ìœ„ ì„¤ì •
plt.show()
```
